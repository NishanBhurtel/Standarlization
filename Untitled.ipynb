```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
```

```python
df.head()
```

```python
df.sample(5)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.drop("Purchased",axis=1),df["Purchased"],test_size=0.3,random_state=0)
```

```python
X_train.shape,X_test.shape
```

## StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# fit the scaler to the train set ,it will learn the parameters
scaler.fit(X_train)

# transfor train and test set 
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)

```

```python
scaler.mean_
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
X_train_scaled
```

```python
 np.round(X_train.describe(),1)
```

```python
 np.round(X_train_scaled.describe(),1)
```

## Effect of scaling


```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(X_train["Age"],X_train["EstimatedSalary"])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled["Age"],X_train["EstimatedSalary"],color="red")
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))

#before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(X_train["Age"],ax=ax1)
sns.kdeplot(X_train["EstimatedSalary"],ax=ax1)

#after scaling
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled["Age"],ax=ax2)
sns.kdeplot(X_train_scaled["EstimatedSalary"],ax=ax2)
```

## Why Scaling is important?


```python
from sklearn.linear_model import LogisticRegression
```

```python
Ir= LogisticRegression()
Ir_scaled=LogisticRegression()
```

```python
Ir.fit(X_train,Y_train)
Ir_scaled.fit(X_train_scaled,Y_train)
```

```python
y_pred=Ir.predict(X_test)
y_pred_scaled=Ir_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual", accuracy_score(Y_test, y_pred))
print("Scaled", accuracy_score(Y_test, y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
dt_scaled= DecisionTreeClassifier()
```

```python
dt.fit(X_train,Y_train)
dt_scaled.fit(X_train_scaled,Y_train)
```

```python
y_pred=dt.predict(X_test)
y_pred_scaled=dt_scaled.predict(X_test_scaled)
```

```python
print("Actual", accuracy_score(Y_test, y_pred))
print("Scaled", accuracy_score(Y_test, y_pred_scaled))
```
